## Part Two

Create two tables of yearly revenue change, one for each of the customer types. Include both the difference and the percent change. 

```python
yearly.head()
#For Education customers
yearly_Education_change= yearly[yearly['CUSTOMER_TYPE']=='Education'].groupby('YEAR').sum()
yearly_Education_change['REVENUE_DIFFERENCE']=yearly_Education_change['REVENUE'].diff()
yearly_Education_change['PERCENTAGE_CHANGE']=yearly_Education_change['REVENUE'].pct_change()
yearly_Education_change
```

```python
#For Tech customers
yearly_Tech_change= yearly[yearly['CUSTOMER_TYPE']=='Tech'].groupby('YEAR').sum()
yearly_Tech_change['REVENUE_DIFFERENCE']=yearly_Tech_change['REVENUE'].diff()
yearly_Tech_change['PERCENTAGE_CHANGE']=yearly_Tech_change['REVENUE'].pct_change()
yearly_Tech_change
```

For each customer type, calculate the total revenue change from adding new customers for each year. This is the sum of REVENUE for NEW customers for each year. You should end up with a series for each customer type.

```python
#For Education customers
yearly_new= yearly[yearly['CUSTOMER_TYPE']=='Education']
yearly_new_data=yearly_new[yearly_new['STATUS']=='NEW']
yearly_new_data=yearly_new_data.loc[:,['YEAR','REVENUE']]
yearly_new_data=yearly_new_data.groupby('YEAR').sum().reset_index()
```

```python
#For Tech customers
yearly_T_new= yearly[yearly['CUSTOMER_TYPE']=='Tech']
yearly_new_Tdata=yearly_T_new[yearly_T_new['STATUS']=='NEW']
yearly_new_Tdata=yearly_new_Tdata.loc[:,['YEAR','REVENUE']]
yearly_new_Tdata=yearly_new_Tdata.groupby('YEAR').sum().reset_index()
```

For each customer type, calculate the total revenue change from existing customers for each year. This is the sum of REVENUE_DIFF for UPSELL and DOWNSELL customers for each year. You should end up with a series for each customer type.

```python
#For Education customers
yearly_new= yearly[yearly['CUSTOMER_TYPE']=='Education']
yearly_existing_data=yearly_new[(yearly_new['STATUS']=='UPSELL')|(yearly_new['STATUS']=='DOWNSELL')]
yearly_existing_data
yearly_existing_data=yearly_existing_data.loc[:,['YEAR','REVENUE_DIFF']]
yearly_existing_data=yearly_existing_data.groupby('YEAR').sum().reset_index()
```

```python
#For Tech customers
yearly_T_new= yearly[yearly['CUSTOMER_TYPE']=='Tech']
yearly_existing_Tdata=yearly_T_new[(yearly_T_new['STATUS']=='UPSELL')|(yearly_new['STATUS']=='DOWNSELL')]
yearly_existing_Tdata
yearly_existing_Tdata=yearly_existing_Tdata.loc[:,['YEAR','REVENUE_DIFF']]
yearly_existing_Tdata=yearly_existing_Tdata.groupby('YEAR').sum().reset_index()
```

Create two stacked bar charts, one for each customer type, that show the revenue change from adding new customers and the revenue change from existing customers.


```python
#For education customers
#yearly_existing_data 
#yearly_new_data
education_merged = pd.concat([yearly_existing_data,yearly_new_data ], ignore_index=False, sort=False, axis=1).iloc[:,[0,1,3]]
education_merged.rename(columns = {'REVENUE':'NEW_CUSTOMER','REVENUE_DIFF':'EXISITNG_CUSTOMERS'}, inplace = True)
alt.Chart(education_merged.melt(id_vars=['YEAR'], var_name='Existing Data Vs New Data', value_name='Revenue Change')).mark_bar(size=10).encode(
    x='YEAR:T',
    y='Revenue Change:Q',
    color='Existing Data Vs New Data:N'
).properties(width=300)
```

```python
#For Tech customers
#yearly_existing_Tdata 
#yearly_new_Tdata
Tech_merged = pd.concat([yearly_existing_Tdata,yearly_new_Tdata], ignore_index=False, sort=False, axis=1).iloc[:,[0,1,3]]
Tech_merged.rename(columns = {'REVENUE':'NEW_CUSTOMER','REVENUE_DIFF':'EXISITNG_CUSTOMERS'}, inplace = True)
alt.Chart(Tech_merged.melt(id_vars=['YEAR'], var_name='Existing Data Vs New Data', value_name='Revenue Change')).mark_bar(size=30).encode(
    x='YEAR:T',
    y='Revenue Change:Q',
    color='Existing Data Vs New Data:N'
).properties(width=300)
```

For each customer type, calculate the yearly revenue lost from customers churning. This is the sum of REVENUE of customers for which CHURN is true. Put these two series into a dataframe.

```python
yearly
#For education customers
yearly_education=yearly[yearly['CUSTOMER_TYPE']=='Education']
yearly_churn_education=yearly_education[yearly_education['CHURN']==True]
yearly_churn_education=yearly_churn_education.loc[:,['YEAR','REVENUE']].groupby('YEAR').sum()
yearly_churn_education.rename(columns = {'REVENUE':'EDUCATION'}, inplace = True)
#yearly_churn_education.reset_index()
#For Tech customers
yearly_education2=yearly[yearly['CUSTOMER_TYPE']=='Tech']
yearly_churn_Tech=yearly_education2[yearly_education2['CHURN']==True]
yearly_churn_Tech=yearly_churn_Tech.loc[:,['YEAR','REVENUE']].groupby('YEAR').sum()
yearly_churn_Tech.rename(columns = {'REVENUE':'Tech'}, inplace = True)
#yearly_churn_Tech.reset_index()
yearly_lost_revenue=pd.concat([yearly_churn_education,yearly_churn_Tech], ignore_index=False, sort=False, axis=1)
yearly_lost_revenue.reset_index(inplace=True)
```

Create a line plot of quarterly revenue change from new customers and existing customers (this line plot should have 2 lines, new and existing).

```python
quarterly
#For New customers
quarterly_new_data=quarterly[quarterly['STATUS']=='NEW']
quarterly_new_data
quarterly_new_data=quarterly_new_data.loc[:,['TRANSACTION_PERIOD','REVENUE']]
quarterly_new_data
quarterly_new_data=quarterly_new_data.groupby('TRANSACTION_PERIOD').sum()
quarterly_new_data.rename(columns={'REVENUE':'New_Revenue'}, inplace=True)
#For existing customers
quarterly_existing_data=quarterly[(quarterly['STATUS']=='DOWNSELL')|(quarterly['STATUS']=='UPSELL')]
quarterly_existing_data
quarterly_existing_data=quarterly_existing_data.loc[:,['TRANSACTION_PERIOD','REVENUE_DIFF']]
quarterly_existing_data=quarterly_existing_data.groupby('TRANSACTION_PERIOD').sum()
quarterly_existing_data.rename(columns={'REVENUE_DIFF':'Existing_Revenue'}, inplace=True)
quarterly_revenue_change=pd.concat([quarterly_new_data,quarterly_existing_data], ignore_index=False, sort=False, axis=1)
quarterly_revenue_change.reset_index(inplace=True)
quarterly_revenue_change
```

## Part Four

The IS_TREATMENT_FLAG column indicates whether a customer received a new promotion. The promotion is designed to increase revenue from the customer.

The promotion occurred in year 2021. So, only in year 2021 do customers have True or False in that column, otherwise it is nan. The value is True for those who received the promotion and False for those who did not.

Using the yearly data, determine by what percentage revenue changed from 2020 to 2021 across the two groups of customers (those who received the promotion and those who did not). 

Hint: Pull out the two groups (True and False) from 2021. For each group, use isin to find the corresponding customers in 2020, and then use isin again to pull out just those customers from 2021. Now you have the 2020 and 2021 values for those customers.

```python
#for those with promotions
yearly_promotions= yearly[yearly['IS_TREATMENT_FLAG']==True]
total_promotion_revenue=yearly_promotions.loc[:,['YEAR','REVENUE']].groupby('YEAR').sum()
total_promotion_revenue.reset_index(inplace=True)
year_2020=yearly[yearly['YEAR']=='2020']
yearly_promotions1=year_2020[year_2020['CUSTOMER_ID'].isin(yearly_promotions['CUSTOMER_ID'])]
yearly_promotions1
yearly_2020_promotions=yearly_promotions1.loc[:,['YEAR','REVENUE']].groupby('YEAR').sum()
yearly_2020_promotions.reset_index(inplace=True)
revenue_change_promotion= (total_promotion_revenue['REVENUE']-yearly_2020_promotions['REVENUE'])/total_promotion_revenue['REVENUE'] *100
print('Revenue change from 2020 to 2021 for those who had promotions: '+ revenue_change_promotion[0].astype(str) +'%')
#for those without promotions
yearly_no_promotions= yearly[yearly['IS_TREATMENT_FLAG']==False]
yearly_no_promotions= yearly_no_promotions[(yearly_no_promotions['YEAR']=='2021')|(yearly_no_promotions['YEAR']=='2020')]
total_no_promotion_revenue=yearly_no_promotions.loc[:,['YEAR','REVENUE']].groupby('YEAR').sum()
total_no_promotion_revenue.reset_index(inplace=True)
year_no_2020=yearly[yearly['YEAR']=='2020']
yearly_no_promotions1=year_no_2020[year_no_2020['CUSTOMER_ID'].isin(yearly_no_promotions['CUSTOMER_ID'])]
yearly_no_promotions1
yearly_no_2020_promotions=yearly_no_promotions1.loc[:,['YEAR','REVENUE']].groupby('YEAR').sum()
yearly_no_2020_promotions.reset_index(inplace=True)
revenue_change_nopromotion= (total_no_promotion_revenue['REVENUE']-yearly_no_2020_promotions['REVENUE'])/total_no_promotion_revenue['REVENUE'] *100
print('Revenue change from 2020 to 2021 for those who had no promotions: '+ revenue_change_nopromotion[0].astype(str) +'%')
```
